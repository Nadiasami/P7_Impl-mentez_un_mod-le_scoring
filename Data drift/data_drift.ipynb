```python
import evidently
import time 
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset, TargetDriftPreset
from evidently import ColumnMapping
import pandas as pd 
import numpy as np
```

```python
df_train = pd.read_csv('C:/Users/smart asus/P7_données/code_vs_code/analyse_nettoyage_experiences/Data_cleaning_for_model/application_train.csv', sep = ",")
df_test = pd.read_csv('C:/Users/smart asus/P7_données/code_vs_code/analyse_nettoyage_experiences/Data_cleaning_for_model/application_test.csv', sep = ",")
```

```python
df_train = df_train.dropna(subset=['TARGET']).drop(columns=['SK_ID_CURR','TARGET'])
df_test = df_test.drop(columns=['SK_ID_CURR'])
```

```python
drift_report = Report(metrics=[DataDriftPreset(), TargetDriftPreset()])
```

```python
drift_report.run(reference_data=df_train, current_data=df_test)
drift_report
```

```python
# Sauvegardez le rapport en tant que fichier HTML
drift_report.save_html('data_drift_report_FULL_script.html')
```
